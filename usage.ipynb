```python
%matplotlib inline
%config InlineBackend.figure_formats = {'png', 'retina'}
import matplotlib.pylab as plt
import pysz_gal.pysz_gal as szgal
import numpy as np
pyszgal_cl = szgal.tsz_gal_cl()
```

```python
ell = np.logspace(1,np.log10(1500),10)

# redshift distrinution of galaxy samples
dndz = np.loadtxt('./dndz_example.txt')

## Set params ##
h0 = 0.67
ombh2 = 0.02225
omch2 = 0.1198
lnAs = 3.065
ns = 0.9645
mnu = 0.06
bias = 1.55
Mcut = 10**11.8
M1 = 10**11.73
kappa = 1.0
sigma_Ncen = 0.15
alp_Nsat = 0.77
rmax = 4.39
rgs = 1.17
z1 = 1e-5
z2 = 0.1

pars = {'h0':h0, 'obh2':ombh2,'och2':omch2,\
        'As':np.exp(lnAs)*1e-10,'ns':ns,'mnu':mnu,\
        'mass_bias':bias,\
        'Mcut':Mcut,'M1':M1,'kappa':kappa,'sigma_Ncen':sigma_Ncen,'alp_Nsat':alp_Nsat,\
        'rmax':rmax,'rgs':rgs,\
        'flag_nu':True,'flag_tll':False}
gg, gy, tll, ng, tmp = pyszgal_cl.get_tsz_cl(ell, pars, dndz,\
                                             z1, z2, z1, z2, 101, 1001)
```

```python
factor = ell*(ell+1)/2/np.pi
plt.loglog(ell,(gg[0]+gg[1])*factor)
plt.loglog(ell,(gg[0])*factor) # 1-halo term
plt.loglog(ell,(gg[1])*factor) # 2-halo term
```
